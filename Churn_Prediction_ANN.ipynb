```python
# importing libraries

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# reading dataset

data = pd.read_csv('/content/drive/MyDrive/Deep Learning dec22/Churn_Modelling.csv')
```

```python
data
```

```python
data.columns
```

```python
# seperating independent and dependent variables

X = data.iloc[:,3:13]
y = data.iloc[:,13]
```

```python
X.info()
```

```python
# converting categorical variable into dummy/indicator variables

geography = pd.get_dummies(X['Geography'], drop_first=True)
gender = pd.get_dummies(X['Gender'], drop_first=True)
```

```python
X = pd.concat([X,geography,gender], axis=1)
```

```python
X.head()
```

```python
# dropping categorical variables post encoding

X.drop(columns=['Geography','Gender'], axis=1, inplace=True)
```

```python
X.head()
```

```python
# splitting data in train and test 

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=101)
```

```python
# creating deep learning model

model = Sequential()

# hidden layer 1
model.add(Dense(200, activation='sigmoid'))

# hidden layer 2
model.add(Dense(150, activation='sigmoid'))

# hidden layer 3
model.add(Dense(250, activation='relu'))

# hidden layer 4
model.add(Dense(100, activation='relu'))

# hidden layer 5
model.add(Dense(50, activation='sigmoid'))

# output layer 
model.add(Dense(1, activation='sigmoid'))
```

```python
# compiling model and checking accuracy

model.compile(optimizer = 'SGD', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# training the model 

model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)
```
